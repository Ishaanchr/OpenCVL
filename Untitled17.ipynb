```python
import numpy as np
import matplotlib.pyplot as plt

triangle = np.array([[0, 0], [1, 0], [0.5, 1], [0, 0]])
def transform(shape, matrix):
    ones = np.ones((shape.shape[0], 1))
    shape_homogeneous = np.hstack([shape, ones])
    transformed = shape_homogeneous @ matrix.T
    return transformed[:, :2]
```

```python
translation_matrix = np.array([[1, 0, 1],
                                [0, 1, 2],
                                [0, 0, 1]])
scaling_matrix = np.array([[2, 0, 0],
                           [0, 2, 0],
                           [0, 0, 1]])
```

```python
theta = np.radians(45)
rotation_matrix = np.array([[np.cos(theta), -np.sin(theta), 0],
                            [np.sin(theta),  np.cos(theta), 0],
                            [0, 0, 1]])
reflection_matrix = np.array([[1, 0, 0],
                               [0, -1, 0],
                               [0, 0, 1]])
shearing_matrix = np.array([[1, 1, 0],
                             [0, 1, 0],
                             [0, 0, 1]])
```

```python
triangle_translated = transform(triangle, translation_matrix)
triangle_scaled = transform(triangle, scaling_matrix)
triangle_rotated = transform(triangle, rotation_matrix)
triangle_reflected = transform(triangle, reflection_matrix)
triangle_sheared = transform(triangle, shearing_matrix)
```

```python
fig, axs = plt.subplots(3, 2, figsize=(12, 12))
axs = axs.flatten()

shapes = [
    (triangle, "Original"),
    (triangle_translated, "Translated"),
    (triangle_scaled, "Scaled"),
    (triangle_rotated, "Rotated"),
    (triangle_reflected, "Reflected"),
    (triangle_sheared, "Sheared"),
]
for ax, (shape, title) in zip(axs, shapes):
    ax.plot(shape[:, 0], shape[:, 1], marker='o')
    ax.set_title(title)
    ax.axis("equal")
    ax.grid(True)

plt.tight_layout()
plt.show()
```
