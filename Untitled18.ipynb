```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = "grills.JPG"
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
def display_image(img, title):
    plt.imshow(img)
    plt.title(title)
    plt.axis("off")
    plt.show()
rows, cols, _ = image.shape
translation_matrix = np.float32([[1, 0, 50], [0, 1, 100]])
translated_image = cv2.warpAffine(image, translation_matrix, (cols, rows))
display_image(translated_image, "Translated Image")

reflected_image = cv2.flip(image, -1)
display_image(reflected_image, "Reflected Image")

center = (cols // 2, rows // 2)
rotation_matrix = cv2.getRotationMatrix2D(center, 45, 1)
rotated_image = cv2.warpAffine(image, rotation_matrix, (cols, rows))
display_image(rotated_image, "Rotated Image")

scaled_image = cv2.resize(image, (cols // 2, rows // 2), interpolation=cv2.INTER_LINEAR)
display_image(scaled_image, "Scaled Image")

M_shear_x = np.float32([[1, 0.5, 0], [0, 1, 0]])
sheared_image_x = cv2.warpAffine(image, M_shear_x, (int(cols * 1.5), rows))
display_image(sheared_image_x, "Sheared Image (X-Axis)")

M_shear_y = np.float32([[1, 0, 0], [0.5, 1, 0]])
sheared_image_y = cv2.warpAffine(image, M_shear_y, (cols, int(rows * 1.5)))
display_image(sheared_image_y, "Sheared Image (Y-Axis)")
```

```python
plt.imshow(image)
plt.title("Original Image - Determine Cropping")
plt.axis("on")
plt.show()

# Get image dimensions
height, width, _ = image.shape
print(f"Image Dimensions: Width = {width}, Height = {height}")

# Approximate crop region for faces (adjust these values based on the output)
# Example: Crop central area of the image
start_row = int(height * 0.2)
end_row = int(height * 0.6)
start_col = int(width * 0.3)
end_col = int(width * 0.7)

# Crop the image
cropped_image = image[start_row:end_row, start_col:end_col]

# Display the cropped image
plt.imshow(cropped_image)
plt.title("Cropped Image - Faces")
plt.axis("off")
plt.show()
```
